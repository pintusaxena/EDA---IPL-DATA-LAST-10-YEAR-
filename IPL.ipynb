```python
def split_train_test(data , test_ration ):
    np.random.seed(42)
    shuffled = np.random.permutation(len(data))
    test_set_size = int(len(data)*test_ration)
    test_indexes = shuffled[:test_set_size]
    train_indexes = shuffled[test_set_size:]
    return data.iloc[train_indexes] , data.iloc[test_indexes]
    
```
